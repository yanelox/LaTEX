```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
 
set_matplotlib_formats ('png')
```

```python
T_eks = list (map (float, input().split()))
T_eks = np.array (T_eks)
```

```python
T_eks * 1000
```

```python
L = 200 * 10 ** (-3)
```

```python
C = list (map (float, input().split()))
C = np.array (C)
C = C * 10 ** (-6)
```

```python
T_teor = 2 * np.pi * np.sqrt (L * C)
T_teor
```

```python
T_teor * 1000
```

```python
fig, ax1 = plt.subplots(figsize=(16, 16), dpi=100)

plt.rc('font', size = 20)

ax1.set_ylabel ("T_, ", size=20) 
ax1.set_xlabel ("T_, ", size=20)

ax1.scatter (T_teor * 1000, T_eks * 1000, marker="+", s = 100)
ax1.plot ([0, T_teor[-1] * 1000], [0, 0.95 * T_eks[-1] * 1000])
ax1.grid(True)

plt.savefig("pictures/graph1.png", dpi=100)
plt.show()
```

```python
R = np.array (list (map (float, input().split())))
R
```

```python
n = np.array (list (map (float, input().split())))
n
```

```python
U_k = np.array (list (map (float, input().split())))
U_k
```

```python
U_kn = np.array (list (map (float, input().split())))
U_kn
```

```python
tt = np.log (U_k / U_kn) / n
tt
```

```python
Rl = 15
```

```python
R_kon = Rl + R
R_kon
```

```python
fig, ax1 = plt.subplots(figsize=(16, 16), dpi=100)

plt.rc('font', size = 20)

ax1.set_xlabel ("1 / R_kont ^ 2,  ^ (-2)", size=20)
ax1.set_ylabel ("1 / tetta^2", size=20)

ax1.scatter (1 / R_kon ** 2, 1 / tt ** 2, marker="+", s = 100)
ax1.plot ([0, 0.85 / R_kon[0] ** 2], [0, 1 / tt[0] ** 2])
ax1.grid(True)

plt.savefig("pictures/graph2.png", dpi=100)
plt.show()
```
