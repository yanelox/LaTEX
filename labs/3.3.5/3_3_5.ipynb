```python
Ec = list(map(float, input().split()))
Ec = np.array (Ec)
Ec = Ec - 12
Ec
```

```python
Ec = Ec / 75 * 3
```

```python
fig, ax1 = plt.subplots(figsize=(16, 16), dpi=100)

ax1.set_title ("     ")

ax1.set_xlabel ("B, ") 
ax1.set_ylabel ("E, ")

ax1.scatter (B, Em1, marker="x", s = 100)
ax1.plot ([B[0], B[-1]], [Em1[0] * 0.3, Em1[-1]], label=", I = 0.6 A")

ax1.scatter (B, Em2, marker="x", s = 100)
ax1.plot ([B[0], B[-1]], [Em2[0], Em2[-1]], label=", I = 0.9 A")

ax1.scatter (B, Em3, marker="x", s = 100)
ax1.plot ([B[0], B[-1]], [Em3[0], Em3[-1]], label=", I = 1.2 A")

plt.legend()
plt.savefig("pictures/graph2.png", dpi=100)
plt.show()
```

```python
km1 = (Em1[-1] - Em1[0] * 0.3) / (B[-1] - B[0])
km2 = (Em2[-1] - Em2[0]) / (B[-1] - B[0])
km3 = (Em3[-1] - Em3[0]) / (B[-1] - B[0])
```

```python
Rxm1 = km1 / 10 ** 3 * 0.09 / 10 ** 3 / 0.6
Rxm2 = km2 / 10 ** 3 * 0.09 / 10 ** 3 / 0.9
Rxm3 = km3 / 10 ** 3 * 0.09 / 10 ** 3 / 1.2
```

```python
print (Rxm1, Rxm2, Rxm3)
```
